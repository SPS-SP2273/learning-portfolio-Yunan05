```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, num=400, endpoint=True)
cos_x = np.cos(x)
sin_x = np.sin(x)
fun1_x = np.exp(-x) * np.cos(5 * x)
fun2_x = np.exp(-x) * np.sin(2 * x)

fig, axes = plt.subplots(
    nrows=2, ncols=2, figsize=(12, 8),
    sharey='row', sharex='col'   
)

# ------- Subplot 1 -------#
axes[0, 0].plot(x, cos_x, color='r', label=r'$\cos x$')
axes[0, 0].plot(x, cos_x**2, color='r', linestyle=':', label=r'$\cos^2 x$')
axes[0, 0].fill_between(x, cos_x, -1, color='r', alpha=.18)   
axes[0, 0].set_title(r'$\cos x$ & $\cos^2 x$')
axes[0, 0].set_ylabel('Cosine Value')
axes[0, 0].legend(loc='upper right')

# ------- Subplot 2 -------#
axes[0, 1].plot(x, sin_x, color='g', label=r'$\sin x$')
axes[0, 1].plot(x, sin_x**2, color='tab:blue', label=r'$\sin^2 x$')
axes[0, 1].fill_between(x, -1, 0, color='g', alpha=.18)      
axes[0, 1].set_title(r'$\sin x$ & $\sin^2 x$')
axes[0, 1].legend(loc='upper right')

# ------- Subplot 3 (Bottom Left) -------#
axes[1, 0].plot(x, fun1_x, color='b', label='sin2x')  
axes[1, 0].fill_between(x, fun1_x, -25, color='b', alpha=.18) 
axes[1, 0].set_title(r'$e^{-x}\cos 5x$')
axes[1, 0].set_xlabel('Angle (radians)')
axes[1, 0].set_ylabel('Cosine Value')
ticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
ticklabels = [r'$-\pi$', r'$-\pi/2$', '0', r'$\pi/2$', r'$\pi$']
axes[1, 0].set_xticks(ticks)
axes[1, 0].set_xticklabels(ticklabels)
axes[1, 0].legend(loc='upper right')

# ------- Subplot 4 (Bottom Right) -------#
axes[1, 1].plot(x, fun2_x, color='y', label='sin2x')  
axes[1, 1].fill_between(x, fun2_x, 0, color='y', alpha=.18)  
axes[1, 1].set_title(r'$e^{-x}\sin 2x$')
axes[1, 1].set_xlabel('Angle (radians)')
axes[1, 1].set_xticks(ticks)
axes[1, 1].set_xticklabels(ticklabels)
axes[1, 1].legend(loc='upper right')

for a in axes.flatten():
    a.grid(alpha=.5)

plt.tight_layout()
plt.savefig("after_subplot_figure.jpg", format="jpg", dpi=150)
plt.show()
```
