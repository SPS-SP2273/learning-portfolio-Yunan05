```python
squares = [i**2 for i in range(10)]
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)
```

```python
%%timeit
squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

Both code snippets produce the same result.

%%timeit shows that list comprehensions are faster than appending in a loop.

The %%time it function helps me to better understand and compare code efficiency.
